```python
class DoublyNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = DoublyNode(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=' <-> ')
            temp = temp.next
        print('None')
```

```python
class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def display(self):
        if not self.head:
            return
        temp = self.head
        while True:
            print(temp.data, end=' -> ')
            temp = temp.next
            if temp == self.head:
                break
        print('(Back to Head)')
```

```python
# Definisi Node dalam Linked List
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name # Menyimpan nama tugas
    self.next = None # Pointer ke tugas berikutnya

# definisi Linked List untuk Daftar Tugas
class TaskList:
  def __init__(self):
    self.head = None # Awalnya tidak ada tugas (daftar kosong)

  # Menambahkan tugas Ke Linked list
  def add_task(self, task_name):
    new_task = TaskNode(task_name)
    if not self.head:
      self.head = new_task # jika List kosong, tugas pertama menjadi head
    else:
      current = self.head
      while current.next: # Mencari Node terakhir dalam List
        current = current.next
      current.next = new_task # Menambahkan tugas di akhir list
  # Menampilkan daftar tugas
  def show_tasks(self):
    if not self.head:
      print("\n Tidak ada tugas hari ini!")
      return
    print("\n Daftar Tugas hari ini:")
    current = self.head
    count = 1
    while current:
      print(f"{count}.{current.task_name}")
      count += 1

# =========== Program Utama (Main) ==============

# Membuat daftar tugas
task_list = TaskList()

# Meminta pengguna memasukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini: "))

# Loop untuk memasukkan tugas ke dalam daftar
for i in range(1, num_tasks + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

# Menampilkan daftar tugas
task_list.show_tasks()
```

```python
# Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url):
    self.url = url # URL halaman WEB
    self.next = None # Pointer ke halaman berikutnya
    self.prev = None #Pointer ke halaman sebelumnya

# definisi Doubly linked List untuk Riwayat Peramban
class BrowserHistory:
  def __init__(self):
    self.current = None # Pointer ke halaman yang terbuka

    # Menambahkan halaman baru
  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page # Jika riwayat kosong, halaman ini menjadi halaman pertama
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page # Pindahkan posisi ke halaman baru

    # Menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n Riwayat kosong!")
      return
    print("\n Riwayat Peramban Web:")
    temp = self.current
    while temp.prev:
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp = temp.next
      count +=1

    # Navigasi mundur ke halaman web
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\n Anda kembali ke: {self.current.url}")
    else:
      print("\n Tidak bisa kembali, Anda berada di halaman pertama.")

    # Navigasi maju ke halaman berikutnya
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.url}")
    else:
      print("\n Tidak bisa maju, Anda berada di halaman terakhir.")

# ========= Program Utama (Main) ==========

# Membuat objek riwayat peramban
history = BrowserHistory() #

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("Masukkan jumlah halaman yang dikunjungi: "))

# Loop untuk memasukkan halaman ke dalam riwayat
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, atau 'x' untuk keluar. \nMasukkan perintah :").lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat peramban.")
    break
  else:
    print('\n Perintah tidak dikenal')
```

```python
# Definisi Node dalam Circular Linked List
class SongNode:
  def __init__(self, title):
    self.title = title # Menyimpan judul lagu
    self.next = None # Pointer ke lagu berikutnya

# Definisi Node untuk Play List lagu
class PlayList:
  def __init__(self):
    self.current = None # Lagu yang sedang diputar

  # Menambahkan lagu ke dalam play list
  def add_song(self, title):
    new_song = SongNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current # Menghubungkan ke dirinya sendiri
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next
      temp.next = new_song # Menambahkan lagu diakhir
      new_song.next = self.current # Menghubungkan kembali diawal

  # Menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print("\n Playlist lagu kosong!")
      return

    print("\n Playlist lagu Anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.title}")
      temp = temp.next
      count += 1
      if temp == self.current:
        print(f"\n Memutar: {self.current.title}")
        break

  # Memutar lagu berikutnya dalam Playlist
  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar: {self.current.title}")
    else:
      print("\n Playlist lagu kosong!")


# ========= Program utama(Main) ===========

# Membuat objek list
playlist = PlayList()

# Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))

# Loop untuk memasukkan lagu ke dalam playlist
for i in range(1, num_songs+1):
  title = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(title)

# Menampilkan daftar lagu dalam playlist
playlist.show_playlist()

# Menu navigasi Pengguna
while True:
  command = input("\n Tekan 'n' untuk lagu berikutnya. 'x' untuk keluar. \nMasukkan perintah: ").lower()

  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print("\n Keluar dari pemutar musik.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("\nPosisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("\nPosisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("\nDaftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.delete_at_beginning()
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp.next:
                print("\nPosisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"\nElemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"\nElemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("\nDaftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# ===== Program Utama (Main) ===== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int((input(f"Masukkan data ke-{i+1}: ")))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n=== Pilih Operasi ===")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balik elemen dalam daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("\nLinked List berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```

```python
class MataKuliahNode:
    def __init__(self, matkul_name):
        self.matkul_name = matkul_name
        self.next = None

class MataKuliahList:
    def __init__(self):
        self.days = {}

    def add_matkul(self, hari, matkul_name):
        if hari not in self.days:
            self.days[hari] = None
        new_matkul = MataKuliahNode(matkul_name)
        if not self.days[hari]:
            self.days[hari] = new_matkul
        else:
            current = self.days[hari]
            while current.next:
                current = current.next
            current.next = new_matkul

    def show_jadwal(self, hari):
        if hari not in self.days or not self.days[hari]:
            print(f"\nTidak ada mata kuliah di hari {hari}!")
            return
        print(f"\nJadwal Mata Kuliah Hari {hari}:")
        current = self.days[hari]
        count = 1
        while current:
            print(f"{count}. Mata Kuliah: {current.matkul_name}")
            current = current.next
            count += 1

def validasi_hari(hari):
    hari_valid = ["Senin", "Selasa", "Rabu", "Kamis", "Jumat", "Sabtu", "Minggu"]
    return hari.capitalize() in hari_valid

def validasi_matkul(matkul_name):
    return len(matkul_name.strip()) > 0

def tampilkan_menu():
    print("\nMenu:")
    print("1. Tambah Mata Kuliah")
    print("2. Lihat Jadwal Mata Kuliah")
    print("3. Keluar")

matkul_list = MataKuliahList()

while True:
    tampilkan_menu()
    try:
        pilihan = int(input("Masukkan pilihan (1-3): "))

        if pilihan == 1:
            matkul_name = input("Masukkan nama mata kuliah: ")
            while not validasi_matkul(matkul_name):
                print("Nama mata kuliah tidak boleh kosong!")
                matkul_name = input("Masukkan nama mata kuliah: ")

            hari = input("Masukkan hari untuk mata kuliah (Senin, Selasa, ...): ")
            while not validasi_hari(hari):
                print("Hari tidak valid! Masukkan hari yang benar (Senin, Selasa, ...).")
                hari = input("Masukkan hari untuk mata kuliah (Senin, Selasa, ...): ")

            matkul_list.add_matkul(hari.capitalize(), matkul_name)
            print(f"Mata kuliah '{matkul_name}' ditambahkan pada hari {hari.capitalize()}.")

        elif pilihan == 2:
            hari_pilih = input("Masukkan hari yang ingin dilihat jadwalnya: ")
            while not validasi_hari(hari_pilih):
                print("Hari tidak valid! Masukkan hari yang benar (Senin, Selasa, ...).")
                hari_pilih = input("Masukkan hari yang ingin dilihat jadwalnya: ")

            matkul_list.show_jadwal(hari_pilih.capitalize())

        elif pilihan == 3:
            print("Terima kasih! Program selesai.")
            break

        else:
            print("Pilihan tidak valid! Silakan pilih antara 1-3.")

    except ValueError:
        print("Masukkan pilihan yang valid (1-3).")
```

```python
class FolderNode:
    def __init__(self, folder_name):
        self.folder_name = folder_name
        self.next = None
        self.prev = None

class FolderHistory:
    def __init__(self):
        self.current = None

    def visit_folder(self, folder_name):
        new_folder = FolderNode(folder_name)
        if not self.current:
            self.current = new_folder
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder

    def show_history(self):
        if not self.current:
            print("\nRiwayat folder kosong!")
            return
        print("\nRiwayat Navigasi Folder:")
        temp = self.current
        while temp.prev:
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.folder_name}")
            temp = temp.next
            count += 1

    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\nAnda kembali ke folder: {self.current.folder_name}")
        else:
            print("\nTidak bisa kembali, Anda berada di folder pertama.")

    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\nAnda maju ke folder: {self.current.folder_name}")
        else:
            print("\nTidak bisa maju, Anda berada di folder terakhir.")

folder_history = FolderHistory()

while True:
    try:
        num_folders = int(input("Masukkan jumlah folder yang dikunjungi: "))
        if num_folders <= 0:
            print("Jumlah folder harus lebih dari 0! Silakan coba lagi.")
        else:
            break
    except ValueError:
        print("Input tidak valid! Harap masukkan angka yang valid.")

for i in range(1, num_folders + 1):
    while True:
        folder_name = input(f"Masukkan nama folder ke-{i}: ").strip()
        if not folder_name:
            print("Nama folder tidak boleh kosong! Silakan coba lagi.")
        else:
            folder_history.visit_folder(folder_name)
            break

folder_history.show_history()

while True:
    command = input("\nTekan 'b' untuk kembali, 'f' untuk maju, atau 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'b':
        folder_history.go_back()
    elif command == 'f':
        folder_history.go_forward()
    elif command == 'x':
        print("\nKeluar dari riwayat navigasi folder.")
        break
    else:
        print("\nPerintah tidak dikenal! Harap masukkan perintah yang valid.")
```
