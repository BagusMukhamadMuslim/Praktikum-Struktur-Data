```python
class Piket_Satpam:
    def __init__(self, nama):
        self.nama = nama
        self.next = None

class Daftar_Piket_Satpam:
    def __init__(self):
        self.head = None

    def tambah_piket(self, nama):
        satpam_baru = Piket_Satpam(nama)
        if not self.head:
            self.head = satpam_baru
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = satpam_baru
            satpam_baru.next = self.head

    def tampilkan_piket(self):
        if not self.head:
            print("\nDaftar satpam kosong!")
            return

        print("\nDaftar satpam:")
        temp = self.head
        count = 1
        while True:
            print(f"{count}. {temp.nama}")
            temp = temp.next
            count += 1
            if temp == self.head:
                break

    def tampilkan_jadwal_piket(self, jumlah_hari):
        if not self.head:
            print("\nDaftar piket satpam kosong!")
            return

        print(f"\nJadwal piket satpam selama {jumlah_hari} hari:")
        temp = self.head
        for hari in range(1, jumlah_hari + 1):
            print(f"Hari {hari}: {temp.nama}")
            temp = temp.next

def perulangan_satpam(daftar_piket):
    while True:
        try:
            num_satpam = int(input("\nMasukkan jumlah satpam: "))
            if num_satpam <= 0:
                print("Jumlah satpam harus lebih dari 0.")
                continue
            break
        except ValueError:
            print("Input harus berupa angka.")

    for i in range(1, num_satpam + 1):
        nama = input(f"Masukkan nama satpam ke-{i}: ").strip()
        if nama:
            daftar_piket.tambah_piket(nama)
        else:
            print("Nama satpam tidak boleh kosong. Ulangi input!")
            i -= 1

daftar_piket = Daftar_Piket_Satpam()

while True:
    print("\nMenu:")
    print("1. Tambah Satpam")
    print("2. Tampilkan Daftar Satpam")
    print("3. Tampilkan Jadwal Piket")
    print("4. Keluar")
    command = input("Masukkan perintah (1-4): ").strip()

    if command == "1":
        perulangan_satpam(daftar_piket)
    elif command == "2":
        daftar_piket.tampilkan_piket()
    elif command == "3":
        while True:
            try:
                jumlah_hari = int(input("Masukkan jumlah hari jadwal piket: "))
                if jumlah_hari <= 0:
                    print("Jumlah hari harus lebih dari 0.")
                    continue
                break
            except ValueError:
                print("Input harus berupa angka.")
        daftar_piket.tampilkan_jadwal_piket(jumlah_hari)
    elif command == "4":
        print("Program selesai.")
        break
    else:
        print("Perintah tidak valid. Coba lagi.")
```
