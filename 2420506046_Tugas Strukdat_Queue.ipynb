```python
class QueueList:
    def __init__(self, max_size=10):
        self.queue = []
        self.max_size = max_size

    def enqueue(self, item):
        if len(self.queue) < self.max_size:
            if "7" in item:
                print(f"Pelanggan VIP customer : {item}.")
                self.queue.insert(0, item)
            else:
                self.queue.append(item)
            return True
        else:
            print("Antrian Penuh!")
            return False

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

queue = QueueList()

while True:
    try:
        print("\n=== ANTRIAN PEMESANAN TIKET BIOSKOP ===")
        print("1. Tambah Pelanggan")
        print("2. Layani Pelanggan")
        print("3. Lihat Antrian")
        print("4. Keluar")
        pilihan = input("Pilih menu: ").strip()

        if pilihan == "1":
            nama = input("Masukkan Nama Pemesan: ").upper().strip()
            npm = input("Masukkan NPM Pemesan: ").strip()
            if npm.isdigit() and len(npm) >= 2:
                ticket_id = nama[:2] + npm[-2:]
                if queue.enqueue(ticket_id):
                    print("Pelanggan", ticket_id, "ditambahkan ke antrian")
                else:
                    print("Antrian sudah penuh, tidak bisa menambahkan pelanggan.")
            else:
                print("NPM harus berupa angka dan memiliki minimal 2 digit.")
        elif pilihan == "2":
            pelanggan = queue.dequeue()
            if pelanggan:
                print("Pelanggan", pelanggan, "telah dilayani.")
            else:
                print("Antrian kosong, tidak ada pelanggan yang dapat dilayani.")
        elif pilihan == "3":
            print("Panjang antrian saat ini:", queue.size() if not queue.is_empty() else "(Kosong)")
            print("Pelanggan pertama:", queue.front())
            print("Pelanggan terakhir:", queue.rear())
        elif pilihan == "4":
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid, coba lagi.")
    except Exception as e:
        print("Terjadi kesalahan:", e)

```

```python
class QueueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, order_id, num_pages):
        if num_pages < 10:
            self.queue.insert(0, (order_id, num_pages))
        else:
            self.queue.append((order_id, num_pages))
        return True

    def dequeue(self):
        if not self.is_empty():
            order_id, pages = self.queue.pop(0)
            if pages > 100:
                self.queue.insert(0, (order_id, pages - 100))
                print(f"Pesanan {order_id} diproses 100 halaman dulu, sisa {pages - 100} halaman.")
            else:
                print(f"Pesanan {order_id} selesai diproses.")
        else:
            print("Antrian kosong!")

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)


queue = QueueList()

while True:
    try:
        print("\n=== ANTRIAN FOTOKOPI KAMPUS ===")
        print("1. Tambah Pesanan")
        print("2. Layani Pesanan")
        print("3. Lihat Antrian")
        print("4. Keluar")
        pilihan = input("Pilih menu: ").strip()

        if pilihan == "1":
            nama = input("Masukkan Nama Pemesan: ").upper().strip()
            npm = input("Masukkan NPM Pemesan: ").strip()
            pages_input = input("Masukkan Jumlah Halaman: ").strip()

            if not npm.isdigit() or len(npm) < 2:
                print("NPM harus berupa angka dan memiliki minimal 2 digit.")
                continue

            if not pages_input.isdigit():
                print("Jumlah halaman harus berupa angka.")
                continue

            pages = int(pages_input)
            order_id = nama[:3] + npm[-2:]

            if queue.enqueue(order_id, pages):
                print(f"Pelanggan {order_id}({pages}) ditambahkan ke antrian")

        elif pilihan == "2":
            queue.dequeue()
        elif pilihan == "3":
            print("Panjang antrian saat ini:", queue.size() if not queue.is_empty() else "(Kosong)")
            print("Pelanggan pertama:", queue.front())
            print("Pelanggan terakhir:", queue.rear())
        elif pilihan == "4":
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid, coba lagi.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

```python
class QueueList:
    def __init__(self, max_size=15):
        self.queue = []
        self.max_size = max_size

    def enqueue(self, item, priority):
        if len(self.queue) < self.max_size:
            if priority == "Electric":
                print(f"Parkir prioritas : {item}")
                self.queue.insert(0, (item, priority))
            else:
                self.queue.append((item, priority))
            return True
        return False

    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        print("Antrian kosong, tidak ada kendaraan untuk dikeluarkan.")
        return None

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

queue = QueueList()

while True:
    try:
        print("\n=== ANTRIAN PARKIR ===")
        print("1. Tambah Kendaraan")
        print("2. Keluarkan Kendaraan")
        print("3. Lihat Parkir")
        print("4. Keluar")
        pilihan = input("Pilih menu: ").strip()

        if pilihan == "1":
            nama = input("Masukkan Kota Lahir Pemilik kendaraan: ").upper().strip()
            npm = input("Masukkan NPM: ").strip()
            priority_input = input("Apakah kendaraan Electric (Y/N): ").strip().upper()

            if not npm.isdigit() or len(npm) < 2:
                print("NPM harus berupa angka dan memiliki minimal 2 digit.")
                continue

            priority = "Electric" if priority_input == "Y" else "Non-Electric"
            vehicle_id = nama[:2] + npm[-2:]

            if queue.enqueue(vehicle_id, priority):
                print(f"Kendaraan {vehicle_id} ({priority}) ditambahkan ke antrian.")
            else:
                print("Kapasitas parkir penuh, tidak bisa menambahkan kendaraan.")

        elif pilihan == "2":
            kendaraan = queue.dequeue()
            if kendaraan:
                print(f"Kendaraan {kendaraan[0]} telah dikeluarkan dari parkiran.")
        elif pilihan == "3":
            print("Panjang antrian saat ini:", queue.size() if not queue.is_empty() else "(Kosong)")
            print("Kendaraan pertama:", queue.front())
            print("Kendaraan terakhir:", queue.rear())
        elif pilihan == "4":
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid, coba lagi.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

```python
class QueueList:
    def __init__(self, max_size=20):
        self.queue = []
        self.regular_queue = []
        self.achievement_queue = []
        self.max_size = max_size

    def enqueue(self, student_id, is_achiever, nasional):
        if nasional:
            self.queue.insert(0, (student_id, is_achiever))
            print(f"Mahasiswa {student_id} berprestasi nasional ditambahkan ke antrian utama di posisi depan.")
        elif len(self.queue) < self.max_size:
            self.queue.append((student_id, is_achiever))
            print(f"Mahasiswa {student_id} ditambahkan ke antrian utama di posisi belakang.")
        else:
            if is_achiever:
                self.achievement_queue.append(student_id)
                print(f"Mahasiswa berprestasi {student_id} mendapat prioritas dan masuk antrean khusus.")
            else:
                self.regular_queue.append(student_id)
                print(f"Mahasiswa {student_id} masuk ke daftar tunggu reguler karena antrean penuh.")

    def dequeue(self):
        if self.queue:
            removed = self.queue.pop(0)
            print(f"Mahasiswa {removed[0]} telah mengambil sertifikat.")

            if self.achievement_queue:
                self.queue.append((self.achievement_queue.pop(0), True))
                print("Mahasiswa berprestasi dari antrean khusus dipindahkan ke antrean utama.")
            elif self.regular_queue:
                self.queue.append((self.regular_queue.pop(0), False))
                print("Mahasiswa reguler dari daftar tunggu dipindahkan ke antrean utama.")
        else:
            print("Antrian utama kosong.")

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def is_empty(self):
        return len(self.queue) == 0 and len(self.achievement_queue) == 0 and len(self.regular_queue) == 0

    def size(self, queue_type):
        if queue_type == "main":
            return len(self.queue)
        elif queue_type == "achievement":
            return len(self.achievement_queue)
        elif queue_type == "regular":
            return len(self.regular_queue)
        else:
            return 0

queue = QueueList()

while True:
    try:
        print("\n=== ANTRIAN PENGAMBILAN SERTIFIKAT ===")
        print("1. Tambah Mahasiswa")
        print("2. Layani Mahasiswa")
        print("3. Lihat Antrian")
        print("4. Keluar")
        pilihan = input("Pilih menu: ").strip()

        if pilihan == "1":
            nama = input("Masukkan Nama Mahasiswa: ").upper().strip()
            npm = input("Masukkan NPM: ").strip()
            is_achiever = input("Apakah mahasiswa berprestasi? (Y/N): ").strip().upper() == "Y"
            nasional = input("Apakah mahasiswa berprestasi nasional? (Y/N): ").strip().upper() == "Y"

            if not npm.isdigit() or len(npm) < 2:
                print("NPM harus berupa angka dan memiliki minimal 2 digit.")
                continue

            student_id = nama[:3] + npm[-2:]
            queue.enqueue(student_id, is_achiever, nasional)

        elif pilihan == "2":
            queue.dequeue()

        elif pilihan == "3":
            print("\n--- STATUS ANTRIAN ---")
            print("Panjang antrian Utama:", queue.size("main") if not queue.is_empty() else "(Kosong)")
            print("Antrian Berprestasi:", queue.size("achievement") if not queue.is_empty() else "(Kosong)")
            print("Antrian Reguler:", queue.size("regular") if not queue.is_empty() else "(Kosong)")
            print("Antrian Pertama:", queue.front())
            print("Antrian Terakhir:", queue.rear())

        elif pilihan == "4":
            print("Keluar dari program.")
            break

        else:
            print("Pilihan tidak valid, coba lagi.")

    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

```
